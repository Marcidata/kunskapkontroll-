1. Förklara övergripande vad NoSQL är.


Svar:

NoSQL är en typ av databas som inte använder traditionella tabeller och rader som i en relationsdatabas. Istället är den mer flexibel och kan hantera stora mängder ostrukturerad data. NoSQL är bra för system som behöver hantera mycket data snabbt, som sociala medier, e-handel och realtidsanalys.

Det finns olika typer av NoSQL-databaser:

Nyckel/värde-databas – Snabb lagring av enkla data.
Dokumentdatabas – Lagrar data i JSON- eller BSON-format.
Bredkolumndatabas – Optimerad för stora dataset.
Grafdatabas – Bra för att hantera relationer mellan data.
Ett exempel på en NoSQL-databas är MongoDB, som används av många företag för att lagra data flexibelt.

2. Vilken typ av data kan en NoSQL-databas hantera som inte en
relationell databas kan?


Svar: 

NoSQL-databaser kan hantera data som:
* Ostrukturerad data – T.ex. inlägg från sociala medier eller loggfiler.
* Hierarkisk data – Lagring av komplexa objekt utan att dela upp dem i flera tabeller.
* Stora datamängder i realtid – Som data från sensorer eller IoT-enheter.
* Relationer mellan data i grafform – T.ex. vänskapsrelationer i sociala medier.

Relationsdatabaser är bra när data är strukturerad och har fasta relationer, men NoSQL är bättre för flexibel och snabb datalagring.

3. Beskriv kortfattat vad en nyckel/värde-databas är.


Svar:

En nyckel/värde-databas lagrar data som ett par av en nyckel och ett värde. Det är som en ordlista där varje nyckel har ett unikt värde kopplat till sig.

Exempel på användning:

Sessionhantering i webbappar – Lagrar användarinformation snabbt.
Caching – Snabb åtkomst till ofta använda data.
Exempel på en nyckel/värde-databas:

Redis – Snabb och enkel datalagring.
Amazon DynamoDB – Används av stora företag.
Exempel på nyckel/värde-lagring:

Nyckel	Värde
"user:1001"	{"namn": "Alice", "ålder": 25}
Det är enkelt och går snabbt att hämta en användare med user:1001.

4. Beskriv kortfattat vad en dokumentdatabas är.


Svar:

En dokumentdatabas lagrar data i JSON- eller BSON-format, vilket gör det enkelt att hantera komplexa objekt utan att dela upp dem i flera tabeller.

Fördelar:
Ingen fast struktur – Kan lagra olika typer av data i samma databas.
Bra för webbapplikationer – Används ofta i moderna system.
Snabb lagring och hämtning – Perfekt för stora datamängder.

Exempel på dokumentdatabaser:

MongoDB – Väldigt populär för webbutveckling.
CouchDB – Bra för offline-synkronisering.
Exempel på ett JSON-dokument i MongoDB:
{
  "_id": "1",
  "namn": "Mobiltelefon",
  "pris": 4999,
  "kategori": "Elektronik"
}
Här lagras man en produkt med namn, pris och kategori i ett JSON-objekt.

5. Beskriv kortfattat vad en bredkolumndatabas är.


Svar:

En bredkolumndatabas sparar data i kolumner istället för rader, vilket gör den mer effektiv för stora dataset. Den används ofta för analys av big data.

Fördelar:
Bra för stora mängder data – T.ex. användardata i stora system.
Optimerad för snabba sökningar – Fungerar bra för analyser.
Används av stora företag – T.ex. Facebook och Google.

Exempel på bredkolumndatabaser:

Apache Cassandra – Skalbar och används av Netflix.
Google Bigtable – Driver Googles sökmotor och Gmail.


6. Beskriv kortfattat vad en grafdatabas är.

Svar:

En grafdatabas lagrar data i noder (objekt) och kanter (relationer). Det gör den perfekt för att analysera samband, t.ex. sociala nätverk eller rekommendationssystem.

Fördelar:
Bra för att hantera relationer – T.ex. vänskapsnätverk.
Snabb data-traversering – Effektivare än JOIN i relationsdatabaser.
Enkel modellering av nätverk – Fungerar som en graf.

Exempel på grafdatabaser:

Neo4j – Används i sociala medier.
ArangoDB – Hybrid mellan graf- och dokumentdatabas.

Ge ett exempel på hur vad man kan använda Redis-databasen du skapat till:

Svar: 

Redis kan användas för att hålla koll på lagersaldo i en t.ex webbshop. När en kund lägger en produkt i sin varukorg och sedan köper den, behöver lagret uppdateras direkt för att säkerställa att ingen annan kan köpa samma produkt om den inte finns kvar. Eftersom Redis är en snabb databas som lagrar information i minnet, kan den uppdatera lagret på några millisekunder. När en produkt köps, minskar Redis automatiskt lagersaldot för den produkten, och om lagret blir slut, kan det direkt visas för kunder att produkten är "slut i lager."

På så sätt slipper kunder se att de kan köpa en produkt, bara för att upptäcka att den inte finns kvar när de försöker betala. Detta gör shoppingupplevelsen smidigare och snabbare, samtidigt som man undviker att sälja produkter som inte finns i lager.
