### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

```python
Det finns två sätt att göra diagram i Matplotlib:
Objektorienterat sätt: Man börjar med att skapa en figur och lägger sedan till axlar.Det ger mer kontroll över hur diagrammet ser ut.

Pyplot-sättet: Man använder funktioner direkt, som plt.plot(), vilket är snabbare och enklare för enklare diagram.

Rekommendation: Om man vill ha mer kontroll över diagrammet, är det bäst att använda det objektorienterade sättet.
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
Figure: Det är hela fönstret eller sidan där diagrammet visas. Tänk på det som ramen för en konstverk.

Axes: Det är den del av figure där själva diagrammet ritas. Det är området där man kan se uppgifterna, inklusive axlarna.

Axis: Detta syftar på axlarna i diagrammet, alltså x-axeln (horisontell) och y-axeln (vertikal) som hjälper till att mäta värden.

Artist: Det handlar om alla synliga delar av diagrammet, som linjer, etiketter och titlar. Man kan säga att en artist skapar det man ser i diagrammet.
```

### When plotting in Matplotlib, what is the expected input data type?

```python
När man plottar i Matplotlib är de vanligaste typerna av indata:

1.Listor eller Arrayer: Man använder ofta listor eller NumPy-arrayer för att mata in data. Till exempel:
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

2.Pandas Series eller DataFrames: Om man jobbar med Pandas kan man använda Series eller DataFrames:

Man kan skapa en DataFrame och sedan plottar:
import pandas as pd
df = pd.DataFrame({'x': [1, 2, 3], 'y': [10, 15, 20]})
plt.plot(df['x'], df['y'])

3.Tuples: Man kan också använda tuples när man vill skapa subplots eller mata in flera dataset.

4.Numeriska värden: Skalar eller andra numeriska värden kan också användas, men de måste passa med hur diagrammet kommer att se ut.

Sammanfattningsvis är listor, arrayer och DataFrames de vanligaste datatyperna som används när man plottar.
Matplotlib är ganska flexibelt och kan hantera olika typer av data för att skapa diagram.
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
# Objektorienterade metoden
Man skapar en figur och axlar först (med plt.subplots()), vilket ger mer kontroll över diagrammet.
Bättre för detaljerade och komplexa diagram.
```

```python
import numpy as np  
import matplotlib.pyplot as plt  

# Skapa x-värden från -4 till 4  
x = np.linspace(-4, 4, 100)  
# Beräkna motsvarande y-värden  
y = x ** 2  

# Skapa en figur och axel  
fig, ax = plt.subplots()  
# Plotta data  
ax.plot(x, y)  

# Sätt titel och etiketter för axlarna  
ax.set_title('Plott av y = x^2')  
ax.set_xlabel('x-värden')  
ax.set_ylabel('y-värden')  

# Visa diagrammet  
plt.show()
```

```python
# Pyplot-metoden
Man plottar direkt med plt.plot(), vilket gör koden kortare och enklare för snabba grafer.
Bättre för snabba och enkla grafer, särskilt om man är nybörjare.
```

```python
import numpy as np  
import matplotlib.pyplot as plt  

# Skapa x-värden från -4 till 4  
x = np.linspace(-4, 4, 100)  
# Beräkna motsvarande y-värden  
y = x ** 2  

# Plotta diagrammet direkt  
plt.plot(x, y)  

# Sätt titel och etiketter för axlarna  
plt.title('Plott av y = x^2')  
plt.xlabel('x-värden')  
plt.ylabel('y-värden')  

# Visa diagrammet  
plt.show()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
# Skapa en figur med 2 underdiagram  
fig, axs = plt.subplots(2, 1, figsize=(8, 10))  # 2 rader, 1 kolumn  

# Spridningsdiagram  
axs[0].scatter(random_data_x, random_data_y, alpha=0.5)  # Rita punkterna  
axs[0].plot(x, y, color='red')  # Rita y = x^2 som en röd linje  
axs[0].set_title('Spridningsdiagram och kurva y = x^2')  # Titel  
axs[0].set_xlabel('x-värden')  # X-axelns titel  
axs[0].set_ylabel('y-värden')  # Y-axelns titel  

# Stapeldiagram  
axs[1].bar(names, values, color='lightblue')  # Rita staplarna  
axs[1].set_title('Frukter och mängd')  # Titel  
axs[1].set_xlabel('Frukt')  # X-axelns titel  
axs[1].set_ylabel('Mängd')  # Y-axelns titel  

# Justera layout  
plt.tight_layout()  

# Visa diagrammet  
plt.show()  
```

```python
För att skapa en figur med två diagram (subplots) använder vi `plt.subplots(2, 1)` för att skapa en layout med två rader och en kolumn. 

- **Första subplotten** visar ett spridningsdiagram med slumpmässiga data och kurvan y = x².
- **Andra subplotten** visar ett stapeldiagram med fruktdatan.

Varje subplot har sin egen titel och axelbeskrivningar. Funktionen `tight_layout()` används för att undvika att diagrammen överlappar, och `show()` används sedan för att visa figuren.

```
